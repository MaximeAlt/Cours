```python
import wikipediaapi
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
wiki = wikipediaapi.Wikipedia('fr')
col = ["Annee", "Titre", "Pays", "Réalisateur", "Entree"]
df = pd.DataFrame()
 
p_list = []
prefix = "Box-office France " 
for i in range(2001,2023):
    nom_p = r"Box-office France " + str(i)
    p_list.append(nom_p)
    

# Iterate through the pages :
for p in p_list :
    annee = p[len(p)-4:]
        
    # Use the Wikipedia API to get the page
    page_url = wiki.page(p)
    
    # Use Beautiful Soup to parse the HTML of the page
    soup = BeautifulSoup(requests.get(page_url.fullurl).content)

    # Find the first tables on the page
    tables = soup.find('tbody')

    # Iterate through the rows of the table
    for row in tables.find_all('tr'):
        cells = row.find_all('td')

        if cells:
            # Get the contents of the cells
            titre = cells[0].text.strip()
            drapeau = cells[1].find_all('img', alt=True)
            pays = []
            for d in drapeau : 
                alt = d['alt']
                clear_alt = alt.replace('Flag of', '').replace('the', '').replace('.svg', '').replace('(1-2)', '').replace(r"d'Amérique", '')
                pays.append(clear_alt)
            réalisateur = cells[2].text.strip()
            entree = cells[3].text.strip().replace('entrées', '') 
            #semaines = cells[4].text.strip()
            agg_row = [annee, titre, pays, réalisateur, entree]
            
           
            #add to the df
            df = pd.concat([df, pd.DataFrame([agg_row], columns=col)], ignore_index=True)

df.to_csv('film.csv')
df.head()
```

```python
wiki = wikipediaapi.Wikipedia('fr')
col = ["Annee", "Titre", "Pays", "Réalisateur", "Entree"]
df = pd.DataFrame()
 
p_list = []
prefix = "Box-office France " 

nom_p = r"Box-office France 2013"

    

# Use the Wikipedia API to get the page
page_url = wiki.page(nom_p)

# Use Beautiful Soup to parse the HTML of the page
soup = BeautifulSoup(requests.get(page_url.fullurl).content)

# Find the first tables on the page
tables = soup.find('tbody')

# Iterate through the rows of the table
for row in tables.find_all('tr'):
    cells = row.find_all('td')

    if cells:
        # Get the contents of the cells
        titre = cells[0].text.strip()
        drapeau = cells[1].find_all('img', alt=True)
        
        # Use a regular expression to extract the desired information
        country_name = re.search('the flag of (.*).svg', drapeau).group(1)
        print(country_name)

        pays = []
        for d in drapeau :
            # Use a regular expression to extract the country name
            match = re.search(r'Flag of (.*)\.svg', drapeau)
            if match:
                country = match.group(1)
                print(country)  # Output: United States
                        
            
            alt = d['alt']
            clear_alt = alt.replace('Flag of', '').replace('the', '').replace('.svg', '').replace('(1-2)', '').replace(r"d'Amérique", '')
            pays.append(clear_alt)
        réalisateur = cells[2].text.strip()
        entree = cells[3].text.strip().replace('entrées', '') 
        #semaines = cells[4].text.strip()
        agg_row = [annee, titre, pays, réalisateur, entree]
        
       
        #add to the df
        df = pd.concat([df, pd.DataFrame([agg_row], columns=col)], ignore_index=True)

df.to_csv('film.csv')
df.head()
```
